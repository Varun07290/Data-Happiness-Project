```python
pop_df = df_2021[['Country name', 'Regional indicator', 'Ladder score',
            'Social support','Healthy life expectancy',
            'Freedom to make life choices', 'Explained by: Generosity','Explained by: Perceptions of corruption']].copy()
```

```python
cols=pop_df.corr()['Ladder score'].sort_values(ascending=False)
fig=plt.figure(figsize=(20,20))
plt.suptitle("Features that contribute to Ladder score(Happiness Index)",family='Serif', weight='bold', size=20)
j=0
for i in cols.index[1:]:
    ax=plt.subplot(421+j)
    ax=sns.regplot(data=pop_df, x='Ladder score',y=i)
    ax.legend('')
    j=j+1
    
plt.legend('')
```
